## Creating and Visualizing Vectors in Numpy

### Definitions  
**scalar** - a number on it's own, not embedded in a vector or a matrix
**vector** - an ordered list of numbers, for example $ \vec{x} = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}$  
**orientation** - whether the vector is laid out horizontally like $\vec{x}$ above, or vertically, such as $\vec{y} = \begin{bmatrix} 4 \\ 5 \\ 6 \\ 7 \end{bmatrix}$ It is convention in linear algebra to assume that all vectors are in column orientation unless otherwise specified.
  
**dimensionality** - how many elements the vector contains. In data science applications we focus on vectors containing only numbers. Dimensionality indicates the dimension of space in which the vector exists. Our example vector $\vec{x}$ above has a dimensionality of 3, so it is in 3-d space. On the other hand, $\vec{y}$, contains 4 elements. Therefore, $vec{y}$ has a dimensionality of 4 and exists in 4-d space. To indicate the dimensionality of a vector containing real numbers, use $\mathbb{R}^N$, where $^N$ is the dimensionality of the vector. So $\vec{x}$ exists is a member of $\mathbb{R}^3$ and $\vec{y}$ is a member of $\mathbb{R}^4$.

**Does Vector Orientation Matter? An aside**  
You might have noticed that $\vec{x}$ and $\vec{z}$ contain the same numbers but are opposite orientations. Is this something that you need to worry about? The answer is, it depends. If you're just using vectors to store data, then orientation usually is inconsequential. However, certain operations depend on using the correct orientation, so if you plan on actually manipulating vectors, it's worth understanding vector orientation and making sure your vectors are oriented correctly.

### Dimensionality in Python ###

Confusingly, in the world of Python, dimensionality has a different meeting. In Python, the dimensionality of a vector refers to the number of geometric dimensions used to print out a numerical object, and *does not depend on the number of elements contained in that vector!*.  
All three vectors above are considered 2D arrays in Python. A list of of numbers without a particular orientation is considered a 1D array in Python. The mathematical dimensionality, that is, the number of elements in a vector, is referred to in Python as the **length** or the **shape**. 


## Creating vectors

```python
# a vector as a Python list datatype. This is the simplest way to represent a vector
asList = [1,2,3]
print("List:", asList)
# many linalg operations won't work on a list though, so it's usually better to create numpy arrays
# same numbers, but as a dimensionless numpy array
asArray = np.array([1,2,3]) # orientationless, just a 1D list of numbers in numpy
print("array:", asArray)

# again same numbers, but now endowed with orientations
# Orientation in numpy is given by brackets: the outermost brackets group all numbers together into one object.
# Then, each additional set of brackets indicates a row
rowVec = np.array([ [1,2,3] ]) # row vector, all numbers in one row
colVec = np.array([ [1],[2],[3] ]) # column, 3 rows with 1 number each
print("row vector:", rowVec)
print("column vector", colVec)

# Note the use of spacing when defining the vectors; that is not necessary but makes the code more readable.
```

```python
# We can explore these orientations by examining the hspaes of the variable
# Check the sizes of the variables
print(f'asList:  {np.shape(asList)}') # using np's shape function
print(f'asArray: {asArray.shape}') # using a method associated with numpy objects
print('2D arrays always have their shape reported as (rows, columns)')
print(f'rowVec:  {rowVec.shape}')
print(f'colVec:  {colVec.shape}')
```

## Geometry of vectors
In addition to the algeraic interpretation of a vector (an ordered list of numbers), there exists a **geometric interpretation of a vector**: a straight line with a specific length (*magnitude*) and direction (*angle*, computed relative eto the positive x-axis). We call the point where a vector starts the *tail*, and the point where it ends the *head*. Unless otherwise specified, you can assume a vector's tail is the origin. In this case, a vector is also encodes a geometric coordinate. The graph below depicts the vector $\vec{v} = \begin{bmatrix} -1 \\ 2 \end{bmatrix}$ in standard position.

```python
# create a vector
v = np.array([-1,2])

# plot that vector (and a dot for the tail)
plt.arrow(0,0,v[0],v[1],head_width=.5,width=.1)
plt.plot(0,0,'ko',markerfacecolor='k',markersize=7)

# add axis lines
plt.plot([-3,3],[0,0],'--',color=[.8,.8,.8],zorder=-1)
plt.plot([0,0],[-3,3],'--',color=[.8,.8,.8],zorder=-1)

# make the plot look nicer
plt.axis('square')
plt.axis([-3,3,-3,3])
plt.xlabel('$v_0$')
plt.ylabel('$v_1$')
plt.title('Vector v in standard position')
plt.show()
```

Below you see a graph of the same vector with various starting positions. Remember, even though the arrows are in different places, they are identical becuase they represent the same vector. In other words, they are the same numbers, and have the same magnitude and direction.

```python
# A range of starting positions

startPos = [
            [0,0],
            [-1,-1],
            [1.5,-2]
            ]


# create a new figure
fig = plt.figure(figsize=(6,6))

for s in startPos:

  # plot that vector (and a dot for the tail)
  # note that plt.arrow automatically adds an offset to the third/fourth inputs
  plt.arrow(s[0],s[1],v[0],v[1],head_width=.5,width=.1,color='black')
  plt.plot(s[0],s[1],'ko',markerfacecolor='k',markersize=7)

  # indicate the vector in its standard position
  if s==[0,0]:
    plt.text(v[0]+.1,v[1]+.2,'"Standard pos."')


# add axis lines
plt.plot([-3,3],[0,0],'--',color=[.8,.8,.8],zorder=-1)
plt.plot([0,0],[-3,3],'--',color=[.8,.8,.8],zorder=-1)

# make the plot look nicer
plt.axis('square')
plt.axis([-3,3,-3,3])
plt.xlabel('$v_0$')
plt.ylabel('$v_1$')
plt.title('Vector $\mathbf{v}$ in various locations')
plt.savefig('Figure_02_01.png',dpi=300) # write out the fig to a file
plt.show()
```

## Vector Operations

### Vector Addition
To add two vectors, simply add each element of one vector to the number in the corresponding space in the other vector.
$$ \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix} + \begin{bmatrix} 10 \\ 20 \\ 30 \end{bmatrix} = \begin{bmatrix} 4+10 \\ 5 + 20 \\ 6 + 30 \end{bmatrix} =  \begin{bmatrix} 14 \\ 25 \\ 36 \end{bmatrix} $$
Obviously, vector subtraction behaves the exact same way.

```python
# Using 2D vectors here instead of 3D vectors in the book to facilitate visualization
v = np.array([1,2])
w = np.array([4,-6])
vPlusW = v+w

# print out all three vectors
print(f'{v} + {w} = {vPlusW}')
# Note that here we don't need to worry about vector orientation (row vs. column), 
# so for simplicity the vectors are created orientationless.
```

How would you add these two vectors?
$$ \begin{bmatrix} 0 \\ 3 \\ 6 \\ 9 \end{bmatrix} + \begin{bmatrix} 10 \\ 20 \\ 30 \end{bmatrix} = \space?? $$
Trick question! Since you add vectors element-wise, each vector must have the same dimensionality. Therefore, you can't add the two vectors above together.  
  
Now, what about these two?
$$ \begin{bmatrix} 0 \\ 3 \\ 6 \\ 9 \end{bmatrix} + \begin{bmatrix} 10 & 20 & 30 & 40\end{bmatrix} =\space?? $$
This looks better. After all, you can add each element right? Except, not really. Since the first vector is a column matrix and the second matrix is a row matrix, the elements positions don't match up exactly.

```python
# in python, attempting to add a row vector to a column vector gets wonky
v = np.array([[4,5,6]])
w = np.array([[10,30,40]]).T # transposing. for now just know that it swaps between a row and a column vector
v + w # this result is super weird due to a feature of numby called broadcasting, which you'll learn more about later
```

### The Geometry of Vector Addition and Subtraction

* When you think about adding two vectors geometrically, think about placing the tail of one vector on the the head of the other vector. The resulting vector from the tail of the first vector to the head of the second vector is the sum of the two vectors. You can repeat extend this to any number of vectors, just continuously stack all the vectors tail to head, and then the sum is the line that goes from the tail of the first vector to the head of the last vector.
* Subtracting vectors is a little different. To subtract two vectors geometrically, place both vectors with their tails at the same coordinates. The difference vector is the line that goes from the head of the 'negative' vector to the to the head of the 'positive' vector. The geometry for vector subtraction is particularly important. It is the basis for orthogonal value decomposition, which in turn is the basis for linear least squares.

### Vector-Scalar Multiplication
To multiply a vector by a scalar, simply multiply each element in the vector by the scalar. For example,
$$ \lambda = 4 $$
$$ \vec{w} = \begin{bmatrix} 9 \\ 4 \\ 1 \end{bmatrix} $$
$$ \lambda\vec{w} = \begin{bmatrix} 4 * 9 \\ 4 * 4 \\ 4 * 1 \end{bmatrix} = \begin{bmatrix} 36 \\ 16 \\ 4 \end{bmatrix}

```python
# Vector-scalar multiplication is our first example where data type really matters. Observe.
s = 2
a = [3,4,5] # a Python list
b = np.array(a) # as a numpy array
print(a*s) # repeats the list s times
print(b*s) # because it's an array, interprets the asterisk as multiplication
```

### The Geometry of Vector-Scalar Multiplication
Scalars change the magnitude of a vector but not it's direction. In the example below, the vector $\vec{b}$, represented by the magenta line, is multiplied by the scalar $s$. The product, $s\vec{v}$, is represented by the black line.

```python
# plot
plt.plot([0,b[0]],[0,b[1]],'m--',linewidth=3,label='b')
plt.plot([0,s*b[0]],[0,s*b[1]],'k:',linewidth=3,label='sb')

plt.grid()
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.legend()
plt.show()
```

The effect of the scalar on the original vector depends on the properties of the scalar. From left to right:
* if the scalar is equal to 1, the vector is unchanged.
* if the scalar is greater than 1, the vector grows in magnitude
* if the scalar is between 0 and 1, the vector shrinks in magnitude
* if the scalar is 0, the vector shrinks down to a point in space.
* if the scalar is negative, the vector is flipped over itself.

```python
# Effects of different scalars

# a list of scalars:
scalars = [ 1, 2, 1/3, 0, -2/3 ]

baseVector = np.array([ .75,1 ])

# create a figure
fig,axs = plt.subplots(1,len(scalars),figsize=(12,3))
i = 0 # axis counter

for s in scalars:

  # compute the scaled vector
  v = s*baseVector

  # plot it
  axs[i].arrow(0,0,baseVector[0],baseVector[1],head_width=.3,width=.1,color='k',length_includes_head=True)
  axs[i].arrow(.1,0,v[0],v[1],head_width=.3,width=.1,color=[.75,.75,.75],length_includes_head=True)
  axs[i].grid(linestyle='--')
  axs[i].axis('square')
  axs[i].axis([-2.5,2.5,-2.5,2.5])
  axs[i].set(xticks=np.arange(-2,3), yticks=np.arange(-2,3))
  axs[i].set_title(f'$\sigma$ = {s:.2f}')
  i+=1 # update axis counter

plt.tight_layout()
plt.savefig('Figure_02_03.png',dpi=300)
plt.show()
```

**A note about negative scalars**
Why did the scalar point in the opposite direction after being multiplied by a negative scalar if scalars don't change the direction of vectors? The answer is based on the fact that vectors lie along an infinitely long line that passes through the origin and continues in both directions (known as "one-dimensional subspace"). Therefore, even though the arrow is pointing in the reverse direction, the vector still lies along this same infinite line. Therefore, it's direction doesn't change.

### Vector Averaging
Averaging vectors is the same as averaging numbers, sum and divide by the number of numbers. For example,
$$ \vec{a} = \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix} $$
$$ \vec{b} = \begin{bmatrix} 10 \\ 20 \\ 30 \end{bmatrix} $$
The average of $\vec{a} and $\vec{b} is therefore:
$$\frac{\begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix} + \begin{bmatrix} 10 \\ 20 \\ 30 \end{bmatrix}}{2} = \begin{bmatrix} 14 \\ 25 \\ 36 \end{bmatrix} * 0.5 = \begin{bmatrix} 7 \\ 12.5 \\ 18 \end{bmatrix}

### Vector-Scalar Addition
Technically, vectors and scalars are two different types of mathematical objects that cannot be combined. Therefore, vector-scalar addition isn't formally defined in linear algebra. However, Python allows vector-scalar addition and treats it like vector-scalar multiplication, adding the scalar to each element of the vector.

```python
v = np.array([1,2])
s = 3.5

print(v)
print(s+v)
```

## Creating and Visualizing Vectors in Numpy

### Definitions  
**scalar** - a number on it's own, not embedded in a vector or a matrix
**vector** - an ordered list of numbers, for example $ \vec{x} = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}$  
**orientation** - whether the vector is laid out horizontally like $\vec{x}$ above, or vertically, such as $\vec{y} = \begin{bmatrix} 4 \\ 5 \\ 6 \\ 7 \end{bmatrix}$ It is convention in linear algebra to assume that all vectors are in column orientation unless otherwise specified.
  
**dimensionality** - how many elements the vector contains. In data science applications we focus on vectors containing only numbers. Dimensionality indicates the dimension of space in which the vector exists. Our example vector $\vec{x}$ above has a dimensionality of 3, so it is in 3-d space. On the other hand, $\vec{y}$, contains 4 elements. Therefore, $\vec{y}$ has a dimensionality of 4 and exists in 4-d space. To indicate the dimensionality of a vector containing real numbers, use $\mathbb{R}^N$, where $^N$ is the dimensionality of the vector. So $\vec{x}$ exists is a member of $\mathbb{R}^3$ and $\vec{y}$ is a member of $\mathbb{R}^4$.

### Dimensionality in Python ###

Confusingly, in the world of Python, dimensionality has a different meeting. In Python, the dimensionality of a vector refers to the number of geometric dimensions used to print out a numerical object, and *does not depend on the number of elements contained in that vector!*.  
All three vectors above are considered 2D arrays in Python. A list of of numbers without a particular orientation is considered a 1D array in Python. The mathematical dimensionality, that is, the number of elements in a vector, is referred to in Python as the **length** or the **shape**.

## Geometry of vectors
In addition to the algeraic interpretation of a vector (an ordered list of numbers), there exists a **geometric interpretation of a vector**: a straight line with a specific length (*magnitude*) and direction (*angle*, computed relative eto the positive x-axis). We call the point where a vector starts the *tail*, and the point where it ends the *head*. Unless otherwise specified, you can assume a vector's tail is the origin. In this case, a vector is also encodes a geometric coordinate. The graph below depicts the vector $\vec{v} = \begin{bmatrix} -1 \\ 2 \end{bmatrix}$ in standard position.

```python
# create a vector
v = np.array([-1,2])

# plot that vector (and a dot for the tail)
plt.arrow(0,0,v[0],v[1],head_width=.5,width=.1)
plt.plot(0,0,'ko',markerfacecolor='k',markersize=7)

# add axis lines
plt.plot([-3,3],[0,0],'--',color=[.8,.8,.8],zorder=-1)
plt.plot([0,0],[-3,3],'--',color=[.8,.8,.8],zorder=-1)

# make the plot look nicer
plt.axis('square')
plt.axis([-3,3,-3,3])
plt.xlabel('$v_0$')
plt.ylabel('$v_1$')
plt.title('Vector v in standard position')
plt.show()
```

Below you see a graph of the same vector with various starting positions. Remember, even though the arrows are in different places, they are identical becuase they represent the same vector. In other words, they are the same numbers, and have the same magnitude and direction.

## Vector Operations

### Vector Addition
To add two vectors, simply add each element of one vector to the number in the corresponding space in the other vector.
$$ \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix} + \begin{bmatrix} 10 \\ 20 \\ 30 \end{bmatrix} = \begin{bmatrix} 4+10 \\ 5 + 20 \\ 6 + 30 \end{bmatrix} =  \begin{bmatrix} 14 \\ 25 \\ 36 \end{bmatrix} $$
Obviously, vector subtraction behaves the exact same way.

```python
# Using 2D vectors here instead of 3D vectors in the book to facilitate visualization
v = np.array([1,2])
w = np.array([4,-6])
vPlusW = v+w

# print out all three vectors
print(f'{v} + {w} = {vPlusW}')
# Note that here we don't need to worry about vector orientation (row vs. column), 
# so for simplicity the vectors are created orientationless.
```

How would you add these two vectors?

Trick question! Since you add vectors element-wise, each vector must have the same dimensionality. Therefore, you can't add the two vectors above together.  
  
Now, what about these two?

This looks better. After all, you can add each element right? Except, not really. Since the first vector is a column matrix and the second matrix is a row matrix, the elements positions don't match up exactly.

### The Geometry of Vector Addition and Subtraction

* When you think about adding two vectors geometrically, think about placing the tail of one vector on the the head of the other vector. The resulting vector from the tail of the first vector to the head of the second vector is the sum of the two vectors. You can repeat extend this to any number of vectors, just continuously stack all the vectors tail to head, and then the sum is the line that goes from the tail of the first vector to the head of the last vector.
* Subtracting vectors is a little different. To subtract two vectors geometrically, place both vectors with their tails at the same coordinates. The difference vector is the line that goes from the head of the 'negative' vector to the to the head of the 'positive' vector. The geometry for vector subtraction is particularly important. It is the basis for orthogonal value decomposition, which in turn is the basis for linear least squares.

### Vector-Scalar Multiplication
To multiply a vector by a scalar, simply multiply each element in the vector by the scalar. For example,
$$ \lambda = 4 $$
$$ \vec{w} = \begin{bmatrix} 9 \\ 4 \\ 1 \end{bmatrix} $$
$$ \lambda\vec{w} = \begin{bmatrix} 4 * 9 \\ 4 * 4 \\ 4 * 1 \end{bmatrix} = \begin{bmatrix} 36 \\ 16 \\ 4 \end{bmatrix}

```python
# Vector-scalar multiplication is our first example where data type really matters. Observe.
s = 2
a = [3,4,5] # a Python list
b = np.array(a) # as a numpy array
print(a*s) # repeats the list s times
print(b*s) # because it's an array, interprets the asterisk as multiplication
```

### The Geometry of Vector-Scalar Multiplication
Scalars change the magnitude of a vector but not it's direction. In the example below, the vector $\vec{b}$, represented by the magenta line, is multiplied by the scalar $s$. The product, $s\vec{v}$, is represented by the black line.

```python
# plot
plt.plot([0,b[0]],[0,b[1]],'m--',linewidth=3,label='b')
plt.plot([0,s*b[0]],[0,s*b[1]],'k:',linewidth=3,label='sb')

plt.grid()
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.legend()
plt.show()
```

The effect of the scalar on the original vector depends on the properties of the scalar. From left to right:
* if the scalar is equal to 1, the vector is unchanged.
* if the scalar is greater than 1, the vector grows in magnitude
* if the scalar is between 0 and 1, the vector shrinks in magnitude
* if the scalar is 0, the vector shrinks down to a point in space.
* if the scalar is negative, the vector is flipped over itself.

```python
# Effects of different scalars

# a list of scalars:
scalars = [ 1, 2, 1/3, 0, -2/3 ]

baseVector = np.array([ .75,1 ])

# create a figure
fig,axs = plt.subplots(1,len(scalars),figsize=(12,3))
i = 0 # axis counter

for s in scalars:

  # compute the scaled vector
  v = s*baseVector

  # plot it
  axs[i].arrow(0,0,baseVector[0],baseVector[1],head_width=.3,width=.1,color='k',length_includes_head=True)
  axs[i].arrow(.1,0,v[0],v[1],head_width=.3,width=.1,color=[.75,.75,.75],length_includes_head=True)
  axs[i].grid(linestyle='--')
  axs[i].axis('square')
  axs[i].axis([-2.5,2.5,-2.5,2.5])
  axs[i].set(xticks=np.arange(-2,3), yticks=np.arange(-2,3))
  axs[i].set_title(f'$\sigma$ = {s:.2f}')
  i+=1 # update axis counter

plt.tight_layout()
plt.savefig('Figure_02_03.png',dpi=300)
plt.show()
```

**A note about negative scalars**
Why did the scalar point in the opposite direction after being multiplied by a negative scalar if scalars don't change the direction of vectors? The answer is based on the fact that vectors lie along an infinitely long line that passes through the origin and continues in both directions (known as "one-dimensional subspace"). Therefore, even though the arrow is pointing in the reverse direction, the vector still lies along this same infinite line. Therefore, it's direction doesn't change.

### Vector Averaging
Averaging vectors is the same as averaging numbers, sum and divide by the number of numbers. For example,
$$ \vec{a} = \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix} $$
$$ \vec{b} = \begin{bmatrix} 10 \\ 20 \\ 30 \end{bmatrix} $$
The average of $\vec{a}$ and $\vec{b}$ is therefore:
$$\frac{\begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix} + \begin{bmatrix} 10 \\ 20 \\ 30 \end{bmatrix}}{2} = \begin{bmatrix} 14 \\ 25 \\ 36 \end{bmatrix} * 0.5 = \begin{bmatrix} 7 \\ 12.5 \\ 18 \end{bmatrix}

### Transpose
Informally, the transpose operation can be said to change row vectors to column vectors, and vice-versa. To formalize this definition to matrices, first we need to know that a matrix has rows and columns, and each element of a matrix has a *(row, column)* index. The transpose swaps those indices, as demonstrated in the below equation:
$$m_{i,j}^T = m_{j,i}$$
To demonstrate applying this formula to vectors, remember that a vector will always have either one row (in the case of a row vector) or one column (in the case of a column vector). So a 6D row vector will have indices $i=1$ and $j \in {1,2,3,4,5,6}$, while a 6D column vector will have indices $i \in {1,2,3,4,5,6}$ and $j=1$. So transposing a row vector into a column vector changes the index that is always equal to one.  
  
An important property is that transposing twice returns the vector or matrix to its original form. In other words,
$$ \vec{v}^{TT} = \vec{v}$$

```python
# Row vector
r = np.array([ [1,2,3] ])

# orientationless array
l = np.array([1,2,3])
```

```python
# print out the vector
print(r), print(' ')

# Transpose the row vector
print(r.T), print(' ')

# double-transpose
print(r.T.T)
```

```python
# Same for the orientationless array
# print the orientationless array
print(l), print(' ')

# Transpose it - wait what? Vectors must have orientation to transpose them!
print(l.T), print(' ')

# double transpose
print(l.T.T)
```

### Vector Broadcasting in Python
Broadcasting exists only in modern computer-based linear alagebra. Essentially, it means to repeat an operation multiple times between one vector and each element of another vector. This allows us to perform arithmetic operations on different shapes without having to explicitly reshape them.

```python
v = np.array([[1,2,3]]).T
w = np.array([[10, 20]])
vw = v + w
print(f'Vector v, with shape {v.shape}:')
print(v)
print('')
print(f'Vector w, with shape {w.shape}:')
print(w)
print('')
print('v + w = ')
print(vw)
print(f'with shape {vw.shape}')
```

```python
print('Repeating the above calculation but with v now a row vector and w a column vector:')
print('')
v = np.array([[1,2,3]])
w = np.array([[10, 20]]).T
vw = v + w
print(f'Vector v, with shape {v.shape}:')
print(v)
print('')
print(f'Vector w, with shape {w.shape}:')
print(w)
print('')
print('v + w = ')
print(vw)
print(f'with shape {vw.shape}')
```

## Vector Magnitude and Unit Vectors
### Magnitude
We've already defined magnitude as the length of the vector, but now we need to talk about calculating it. Formally, the magnitude is the length of a vector from tail to head and is computed using the standard Euclidean norm (the $_2$ may be implied by convention):
$$\|\vec{v}\|_2 = \sqrt{\sum_{i=1}^{n} v_i^2}$$
  
It's time for another aside about the difference between abstract Linear Algebra and Python linear algebra. In traditional linear algebra, the *dimensionality* of a vector is the number of elements in that vector, while length is the geometric distance. In Python, however, the function `len()` will return the dimensionality, even though it is short for length. To calculate the mathematical definition of length, you can use the function `np.norm()`.



```python
# dimensionality vs magnitude in python
v = np.array([1,2,3,7,8,9])
v_dim = len(v)
v_mag = np.linalg.norm(v)
print(
    f'The dimensionality of v, given by the function len(v) is {v_dim} '
    f',which you might also recognize from the shape of v, {v.shape}.'
)
print(f'The magnitude, or geometric length of v or norm of v is calculated using np.linalg.norm, which returns {v_mag}.')
```

### Unit Vectors
A *unit vector* is a vector with a magnitude of 1 ($\|\vec{v}\|_2 = 1$). Every nonunit vector has an associated unit vector. To create the unit vector from a nonunit vector, scale the nonunit vector by the reciprocal of the vector's norm, like so:

$$\hat{v} =\frac{1}{\|\vec{v}\|_2}\vec{v} = \frac{\vec{v}}{\|\vec{v}\|_2}$$

## The Vector Dot Product

The dot product (also called the inner product) is a single number that can be interpreted as a measure of similary or mapping between two vectors. It can be written as $\vec{a}^{T}\vec{b}$, $\vec{a}\cdot\vec{b}$, or $\langle a, b \rangle$. To calculate the dot product, use the formula:
$$\delta = \sum_{i=1}^{n}a_ib_i$$
  
In English, you multiply each element of $\vec{a}$ with the corresponding element of $\vec{b}$, then add up the products, as seen in the below example.
$$\begin{bmatrix} 1 \\ 2 \\ 3 \\ 4 \end{bmatrix} \cdot \begin{bmatrix} 5 \\ 6 \\ 7 \\ 8 \end{bmatrix} = (1*5)+(2*6)+(3*7)+(4*8) = 5 + 12 + 21 + 32 = 70$$

A dot product can only be calculated between two vectors of the same dimensionality.

The dot product is both commutative $\vec{a}\cdot\vec{b}$ = $\vec{b}\cdot\vec{a}$ and distributive, $\vec{a}\cdot(\vec{b} + \vec{c}) = (\vec{a}\cdot\vec{b}) + (\vec{a}\cdot\vec{c})$

```python
# implementing the dot product in python
v = np.array([1, 2, 3, 4])
w = np.array([5, 6, 7, 8])
print(np.dot(v,w)) #np.dot is the most straightward way
print(np.dot(w,v)) # commutative!
```

```python
# scaling one vector scales the dot product by the same amount
s = 10
r = 0.5
t = -2
print(np.dot(s*v, w))
print(np.dot(r*v, w))
print(np.dot(t*v, w))
```

```python
# just some random vectors...
v = np.array([ 0,1,2 ])
w = np.array([ 3,5,8 ])
u = np.array([ 13,21,34 ])

# two ways to comptue
res1 = np.dot( v, w+u )
res2 = np.dot( v,w ) + np.dot( v,u )

# show that they are equivalent
res1,res2 # the dot product is also distributive!
```

### Interpreting the Dot Product

The simplest way to interpet the dot product is that the larger the dot product, the more similar the two vectors are. Of course, because the magnitude of the dot product depends on the scale of the data, this is only really useful when comparing two vectors of the same data. In other words, the dot product of two vectors containing data measured in grams and centimeters would be larger than the same data measured in pounds and feet. To be broadly useful, therefore, a scaling factor must be introduced. A common scaling factor is the Pearson correlation coefficient, which you'll see again later.

### Geometric Definition of the Dot Product
From a geometric perspective, the dot product is defined as the product of the magnitude of two vectors, scaled by the cosine of the angle between them. In math notation, 
$$ \alpha = \cos(\theta_{\vec{v},\vec{w}})\|\vec{v}\|\|\vec{w}\| $$

This equation and the equation in the first definition are equiavlent (the proof is outside the scope)

### Some More Dot Product Properties

Vector magnitudes are strictly positive quantities. Therefore, the sign of the dot product is determined entirely by the angle between the two vectors.
* *Acute angle*: $\cos(\theta) < 90^{\circ} \rightarrow \alpha > 0 $
* *Obtuse angle*: $\cos(\theta) > 90^{\circ} \rightarrow \alpha < 0 $
* *Collinear in the same direction*: $ \cos(\theta) = 1 \rightarrow \alpha = \|\vec{a}\|\|\vec{b}\| $
* *Collinear in the opposite directions*: $ \cos(\theta) = -1 \rightarrow \alpha = -\|\vec{a}\|\|\vec{b}\| $

**Most importantly**

* *Orthogonal vectors*: $ \cos(\theta) = 90^{\circ} \rightarrow \alpha = 0$
 

## Other Vector Multiplications

### Hadamard Multiplication
Element-wise multiplication of vectors.
$$ \begin{bmatrix} 5 \\ 4 \\ 8 \\ 2 \end{bmatrix} \odot \begin{bmatrix} 1 \\ 0 \\ 0.5 \\ -1 \end{bmatrix} = \begin{bmatrix} 5 * 1 \\ 4 * 0 \\ 8 * 0.5 \\ 2 * -1 \end{bmatrix} = \begin{bmatrix} 5 \\ 0 \\ 4 \\ -2 \end{bmatrix} $$

```python
# Hadamard Multiplication in Python
#a = np.array([5,4,8,2])
#b = np.array([1,0,0.5])
#print(a*b)

# In the above format you get this error:
# ValueError: operands could not be broadcast together with shapes (4,) (3,) 
# This is because Hadamard products are only defined for fectors with the same dimensionality. Try again below.

a = np.array([5,4,8,2])
b = np.array([1,0,0.5, -1])
hadamard = a * b
print(hadamard)
```

### Hadamard Multiplication vs Dot Product
This is best understood with an example. If you have data on number of widgets sold in different shops, and the price per widget in each shop. If you represented the number of widgets as one vector and the price per widget as another vector, the Hadamard Multiplication of the two vectors would give you a vector that would contain the widget revenue *per shop* while the dot product would result in a scalar giving you the total revenue across all shops.

```python
# Notice that the dot product of a and b is the sum of the elements of the Hadamard product.
# In other words, Hadamard multiplication is the first step in computing the dot product!
print('Hadamard product of a and b)')
print(hadamard)
print('sum of elements of hadamard')
print(np.sum(hadamard))
print('dot product of a and b')
print(np.dot(a,b))

```

### Outer Product
The outer product creates a matrix from a column vector and a row vector, seen below.

$$ \begin{bmatrix} a \\ b \\ c \end{bmatrix} \begin{bmatrix} d & e \end{bmatrix} = \begin{bmatrix} ad & ae \\ bd & be \\ cd & ce \end{bmatrix} $$

There are two ways to think about this.
1. Each row in the outer product matrix is the row vector scaled by the corresponding element in the column vector.
2. Each column in the outer product matrix is the column vector scaled by the corresponding element in the row vector.

The outer product is essentially matrix multiplication of two matrices each with a dimension of m, and so the properties of computing it will be discussed in the chapter on matrix multiplication. For now, we'll show an example in math notation and an example in numpy.

$$\begin{bmatrix} 5 \\ 4 \\ 8 \\ 2 \end{bmatrix} \begin{bmatrix} 1 & 0 & 0.5 & -1 \end{bmatrix} = \begin{bmatrix} 5*1 & 5*0 & 5*0.5 & 5*-1 \\ 4*1 & 4*0 & 4*0.5 & 4*-1 \\ 8*1 & 8*0 & 8*0.5 & 8*-1 \\ 2*1 & 2*0 & 2*0.5 & 2*-1 \end{bmatrix} = \begin{bmatrix} 5 & 0 & 2.5 & -5 \\ 4 & 0 & 2 & -4 \\ 8 & 0 & 4 & -8 \\ 2 & 0 & 1 & -2 \end{bmatrix}$$

```python
# outer product in numpy
a = np.array([5,4,8,2])
b = np.array([1,0,0.5, -1])
print('using np.outer:')
print(np.outer(a,b))
print('works regardless of order')
print(np.outer(b,a))
print('')
# numpy can also product the outer product using np.dot if the vectors are in, respectively, row and column orientation
# changing vector a to row orientation
a = np.array([[5, 4, 8, 2]])
print(f'vector a with shape {a.shape}')
print(a)
b = np.array([[1], [0], [0.5], [-1]])
print(f'vector b with shape {b.shape}')
print(b)
print('using np.dot (b,a):')
print(np.dot(b,a))
print('using np.dot(a,b)')
print(np.dot(a,b))
```

## Orthogonal Vector Decomposition

To "decompose" a matrix or a vector means to break it up into multiple simpler pieces. For example, we can decompose 42.01 into two pieces: 42 and .01. This might be because .01 is just noise we can ignore or to save on memory. Either way, we now represent one object (42.01) as the sum of simpler objects (42 + .01). Let's say we decided to ignore the .01, leaving us with 42. We can further decompose 42 into the product of the prime numbers 2, 3, and 7. Again, we have broken down one object into smaller, simpler objects (this is called *prime factorization* and has applications in  numerical processing and cryptography).

## Orthogonal Vector Decomposition - Introduction
Orthogonal Vector Decomposition breaks up a vector into 2 separate vectors, one of which is orthogonal to a reference vector, and another of which is parallel to that same reference vector. Walking through an example is helpful.

Suppose we have 2 vectors, $\vec{a} = (4,7)$ and $\vec{b} = (5,3)$ in standard position.

```python
# Ignore this code on your first read through, it exists to set up variables for the graph we will construct
a = np.array([4, 7])
v = np.array([5, 3]) - 0.63076923 * a   # b - a

a_hat = a / np.linalg.norm(a)
v_hat = v / np.linalg.norm(v)

s = 0.25

P = np.array([2.52307692, 4.41538462])

p1 = P + s * a_hat
p2 = p1 + s * v_hat
p3 = P + s * v_hat
```

Now, suppose we want to find the point on $\vec{a}$ that is as close as possible to the head of $\vec{b}$. Expressed differently, projecct $\vec{b}$ onto $\vec{a}$ such that the projection distance is minimized. Because that point, will be on $\vec{a}$, it will represent a vector that is a scaled version of. Let $\beta$ represent this scalar, so that $\beta\vec{a}$ represents the point of this projection.  

Now that we have 3 vectors, $\vec{a}$, $\vec{b}$, and $\beta\vec{a}$, we can define the line from $\vec{b}$ to $\beta\vec{a}$ using vector subtraction: $\vec{b} - \beta\vec{a}$. (Why? Because the difference vector is the line that goes from the head of the 'negative vector' - in this case, $\beta\vec{a}$ - to the head of the 'positive' vector - in this case $\vec{b}$).

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot an arrow starting at (1, 1) and ending at (3, 4)
# x=1, y=1 (start point)
# dx=2, dy=3 (length/direction: 3-1=2, 4-1=3)
ax.arrow(0,0,4,7, head_width=0.1, head_length=0.2, fc='blue', ec='blue')
ax.arrow(0,0,5,3, head_width=0.1, head_length=0.2, fc='red', ec='red')
ax.text(4.1, 7.3, r'$\vec{a}$', color='blue', fontsize=12)
ax.text(5.3, 3.1, r'$\vec{b}$', color='red', fontsize=12)
ax.plot([2.52307692,5], [4.41538462,3], color = 'black', linestyle = '--')
ax.text(3.25,4.25, r'$\vec{b}-\beta\vec{a}$',color = 'black', fontsize = 12)
ax.plot(2.52307692, 4.41538462, 'ko')
ax.text(2, 4.5, r'$\beta\vec{a}$', color = 'black', fontsize = 12)
ax.plot([P[0], p1[0]], [P[1], p1[1]], color='black')
ax.plot([p1[0], p2[0]], [p1[1], p2[1]], color='black')
ax.plot([p2[0], p3[0]], [p2[1], p3[1]], color='black')

# Set limits and labels for clarity
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

plt.show()
```

If we assume that the line $\vec{b}-\beta\vec{a}$ does indeed intersect with $\vec{a}$ at the point closest to the head of $\vec{b}$ (which we can, because I actually did the math in order to draw that plot), we are able to derive a key insight from that picture: The line that forms the shortest distance between the head of $\vec{b}$ and $\vec{a}$ ($\vec{b}-\beta\vec{a}$), will always be the line from the head of $\vec{b}$ that intersects with $\vec{a}$ at a right angle! For a demonstration of this, look at the plots below. Notice the length of lines $c$ and $d$ are both longer than $\beta\vec{a}$.

```python
# Create a figure and axes
fig, ax = plt.subplots()

# elements of plot

#vector a
ax.arrow(0,0,4,7, head_width=0.1, head_length=0.2, fc='blue', ec='blue')
ax.text(4.1, 7.3, r'$\vec{a}$', color='blue', fontsize=12)
# vector b
ax.arrow(0,0,5,3, head_width=0.1, head_length=0.2, fc='red', ec='red')
ax.text(5.3, 3.1, r'$\vec{b}$', color='red', fontsize=12)

# intersecting line, b - Ba
ax.plot([2.52307692,5], [4.41538462,3], color = 'black', linestyle = '--')
ax.text(3.25,4.25, r'$\vec{b}-\beta\vec{a}$',color = 'black', fontsize = 12)

# point Ba, point at which line from head of b intersects a at the shortest distance
ax.plot(2.52307692, 4.41538462, 'ko')
ax.text(2, 4.5, r'$\beta\vec{a}$', color = 'black', fontsize = 12)

# lines to demonstrate that the norm gets larger as the angle moves farther from 90 degrees
ax.plot([1.15,5], [2,3], color = 'magenta', linestyle = '--')
ax.text(2,2.5, 'c',color = 'magenta', fontsize = 12)
ax.plot([4,5], [7,3], color = 'green', linestyle = '--')
ax.text(4.6,5, 'd',color = 'green', fontsize = 12)

# right angle squares
ax.plot([P[0], p1[0]], [P[1], p1[1]], color='black')
ax.plot([p1[0], p2[0]], [p1[1], p2[1]], color='black')
ax.plot([p2[0], p3[0]], [p2[1], p3[1]], color='black')

# Set limits and labels for clarity
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

plt.show()
```

Now, remember a special property that orthogonal vectors have - their dot product is 0! So, in this case:
$$\vec{a} \cdot (b - \beta\vec{a}) = 0$$

From here, it's just algebra to solve for $\beta$ (note that we're using the other notation for dot product, $\vec{a}^T(\vec{b} - \beta\vec{a})$:

$$
\begin{align}
\vec{a}^T(\vec{b} - \beta\vec{a}) &= 0 \\\
\vec{a}^T\vec{b} - \vec{a}^T\beta\vec{a} &= 0 \\
- \vec{a}^T\beta\vec{a} &= - \vec{a}^T\vec{b} \\
\vec{a}^T\beta\vec{a} &= \vec{a}^T\vec{b} \\
\beta &= \frac{\vec{a}^T\vec{b}} {\vec{a}^T\vec{a}}
\end{align}
$$

The resulting formula is called *orthogonal projection*, and it is the necessary grounding for orthogonal vector decomposition (not to mention the basis for applications in stats and ML, as you'll see later).


### Moving on to Orthogonal Vector Deomposition
Now that you understand the minimum distance projection, it's time to move on to orthogonal vector decomposition.   

**The setup**: you begin with 2 vectors, a target vector $\vec{t}$ and a reference vector $\vec{r}$.  
**The goal**: to decompose the target vector into two other vectors such that (1) those two vectors sum to the target vector and (2) one vector is orthogonal to the reference vector while the other is parallel to the reference vector. We'll call the orthogonal vector $\vec{t_{\perp r}}$ and the parallel vector  $\vec{t_{\parallel r}}$.  

The graph below sets up the problem visually.

```python
# Create a figure and axes
fig, ax = plt.subplots()

# elements of plot

#vector t
ax.arrow(1,1,2,2, head_width=0.1, head_length=0.2, fc='black', ec='black')
ax.text(3.1, 3.3, r'$\vec{t}$', color='black', fontsize=12)

# vector r
ax.arrow(1,1,3,0, head_width=0.1, head_length=0.2, fc='grey', ec='grey')
ax.text(4.25, 0.9, r'$\vec{r}$', color='grey', fontsize=12)

# vector perpendicular to reference vector
ax.arrow(1,1,0,2, head_width=0.1, head_length=0.2, fc='black', ec='black')
ax.text(0.9,4.3, r'$\vec{t_{\perp r}}$',color = 'black', fontsize = 12)

# vecter parallel to reference vector
ax.arrow(1,1,2,0, head_width=0.1, head_length=0.2, fc='black', ec='black')
ax.text(2.45, 0.6, r'$\vec{t_{\parallel r}}$', color='black', fontsize=12)

# lines indicating that the two vectors you're finding sum to the target vector: 
ax.plot([1,3], [3,3], color = 'black', linestyle = '--')
ax.plot([3,3], [1,3], color = 'black', linestyle = '--')
# Set limits and labels for clarity
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)

plt.title('Visual Representation of Orthonal Vector Decomposition')
ax.text(2, 4.5, r'Goal: Given, $\vec{t}$ and $\vec{r}$, find $\vec{t_{\perp r}}$ and  $\vec{t_{\parallel r}}$')
plt.show()
```

**Step 1: Define the parallel component**  
Any scaled version of $\vec{r}$ will be parallel to $\vec{r}$. Therefore, we can apply the orthogonal projection formula we just discovered to find $\vec{t_{\parallel r}}$.  
 Note that rather just trying to find the scalar $\beta$, we are trying to find the scaled version of $\vec{r}$, $\vec{t_{\parallel r}}$.   
 Since $\vec{t_{\parallel r}} = \beta\vec{r}$ and $\beta = \frac{\vec{a}^T\vec{b}} {\vec{a}^T\vec{a}}$, $\vec{t_{\parallel r}} = \frac{\vec{t}^T\vec{r}}{\vec{r}^T\vec{r}} * \vec{r}$, or, written more nicely, the formula below.
$$ \vec{t_{\parallel r}} = \vec{r}\frac{\vec{t}^T\vec{r}}{\vec{r}^T\vec{r}}$$



**Step 2: Define the perpindicular component**  
We know given the problem setup that $\vec{t_{\perp r}} + \vec{t_{\parallel r}} = \vec{t}$. We were given $\vec{t}$ in the problem setup and just found $\vec{t_{\parallel r}}$ in the previous step. Therefore, sipmle algebra gives us 
$$ \vec{t_{\perp r}} = \vec{t} - \vec{t_{\parallel r}} $$

**Bonus Step from Chandler's brain: Why does the orthogonal projection formula show up? Why is it necessary that the distance between $\vec{t}$ and $\vec{t_{\parallel r}}$ be minimized?**

*Answer*
This isn't a condition imposed by the problem, but it follows from the math of the problem setup.

You are solving the equation $$\vec{t} = \vec{t_{\parallel r}} + \vec{t_{\perp r}}$$

Such that
1. $\vec{t_{\parallel r}}$ is parallel to $\vec{r}$, and
2. $\vec{t_{\perp r}}$ is orthogonal to $\vec{r}$

If condition 1 holds, we saw above that $\vec{t_{\parallel r}}$ must equal a scaled version of $\vec{r}$.
$$\vec{t_{\parallel r}} = \alpha\vec{r}$$
for some scalar $\alpha$.  

Rewriting the equation you're trying to solve:

$$
\begin{align}
\vec{t} &= \vec{t_{\parallel r}} + \vec{t_{\perp r}} \\
\vec{t} &= \alpha\vec{r} + \vec{t_{\perp r}} \\
\vec{t_{\perp r}} &= \vec{t} - \alpha\vec{r}
\end{align}
$$

Impose condition 2:
$$
\begin{align}
\vec{t_{\perp r}} \cdot \vec{r} &= 0 \\
\vec{t} - \alpha\vec{r} \cdot \vec{r} &= 0
\end{align}
$$

And solve to find the scalar $\alpha$:

$$
\begin{align}
\vec{t_{\perp r}} \cdot \vec{r} &= 0 \\
(\vec{t} - \alpha\vec{r}) \cdot \vec{r} &= 0 \\
\vec{t} \cdot \vec{r} - \alpha\vec{r} \cdot \vec{r} &= 0 \\
\alpha\vec{r} \cdot \vec{r} &= \vec{t} \cdot \vec{r} \\
\end{align}
$$

Factoring $\alpha$ out of the left hand term is allowed due to the property above - scaling a vector scales its dot product by the same amount. In other words, $(\alpha\vec{r}) \cdot \vec{r} = \alpha(\vec{r} \cdot \vec{r})$.

$$
\begin{align}
\alpha\vec{r} \cdot \vec{r} &= \vec{t} \cdot \vec{r} \\
\alpha(\vec{r} \cdot \vec{r}) &= \vec{t} \cdot \vec{r} \\
\alpha &= \frac{\vec{t} \cdot \vec{r}}{\vec{r} \cdot \vec{r}}
\end{align}
$$

This formula certainly looks familiar. Remember, $\beta = \frac{\vec{a}^T\vec{b}} {\vec{a}^T\vec{a}}$ and $ \vec{t_{\parallel r}} = \vec{r}\frac{\vec{t}^T\vec{r}}{\vec{r}^T\vec{r}}$.

In other words, the distance minimization isn't a given constraint of the problem, it is a condition of enforcing orthogonality. Among all vectors parallel to $\vec{r}$, the one that is whose diffference from $\vec{t}$ is orthogonal to $\vec{r}$ is necessarily the one with the minimum distance between the head of $\vec{r}$ and $\vec{t}$.

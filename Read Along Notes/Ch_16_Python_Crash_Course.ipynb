**Functions**

Functions generally follow the format ouput = function_name(input)

**Methods**  
A *method* is a function that is called directly on a variable. Different data types have different methods associated with them, so a method that works on lists may not work on strings. To call a method, you use a period after the variable on which you're calling the method: variable.method(input).

```python
# writing your own functions
def add2numbers(n1, n2):
    total = n1 + n2
    print(total)
    return total
```

```python
s = add2numbers(4,5)
print(s)
```

**Libraries**

```python
# 1st important library - numpy
import numpy as np
```

```python
# some basic numpy functions
avg = np.mean([1,2,3]) # computes the average
sorted1 = np.sort([2,1,3]) # sorts from lowest to highest
theRank = np.linalg.matrix_rank([[1,2], [1,3]]) # linlalg is a module of numpy containing functions specific to linear algebra, don't worry about matrix rank for now
print(avg)
print(sorted1)
print(theRank)
```

```python
# Numpy has it's own data type called an array
# arrays are like lists, but they can store only numbers and have additional attributes that make them suitable for mathematical coding
vector = np.array([9, 8, 1, 2])
print(f'The variable vector is of type {type(vector)} and looks like {vector}')
```

```python
arr = np.arange(-5, 5) # creates a numpy array from -5 to 4. Note that the first number is inclusive, but the second number is exclusive
print(type(arr))
print(arr)
# to select a certain element from the array, you can index, similar to a list
print(arr[5])
# to select multiple elements you move from indexing to slicing
# to slice, specify the start and end indices with a colon in between, keeping in mind the exlusive upper bound used on Python
# example: get the first 3 elements of our array
print(arr[0:4]) # slice up to 3 + 1 = 4
print(arr[:4]) # if you leave the starting or ending number blank it will start at the beginning or go to the end, depending on which you leave blanks
# you can also skip numbers using the formulation [start:stop:skip]
print(arr[0::2]) # prints every second element - notice I skipped the end of the array so it went all the way to the end
print(arr[::-1]) # -1 in the skipping place prints out the whole array backwards
```

Below we introduce subplots. A subplot is a way of dividing the graphics area (called the figure) into a grid of spearate axes into which can you draw different visualizations.

```python
_,axs = plt.subplots(1,2,figsize=(8,5)) # creates subplots - 1 row of 2 and determines the total figure size as 8x5. since it's assigned to a variable, a tuple is created with the first element being the figure and the second element being a numpy array of axis objects
# the _ is a conventional placeholder indicating that the fig variable is intentionally ignored
np.random.seed(42) # setting seed so random functions below always produce the same result
lines = np.random.randn(10,5) # creates a numpy array with 10 rows and 5 columns full of random numbers drawn from a Gaussian distribution
axs[0].plot(lines) # indexes the first subplot created above (axis[0]). It treates each column in lines as a separate line, so you get 5 lines, each of which hits all 10 points in the respective column. Try to trace the lines.
```

```python
# let's show just one line's points and the line itself to trace it to demonstrate what's going on
blue_line = lines[:,0]
print(blue_line)
_,axs = plt.subplots(1,2,figsize=(8,5))
axs[0].plot(blue_line)
```

```python
_,axs = plt.subplots(1,2,figsize=(8,5))
axs[0].plot(lines)
axs[1].imshow(lines) # visualizes an array as a heatmap. By default, darker colors use lower values.
print(lines) # compare the matrix and the heatmap! the heatmap is a 10x5 grid as well, and each square represents a value in our matrix. 
```

Generate the following equation into code and create a plot.  
$$ f(x) = \frac{\alpha}{1 + e^{-\beta x}} $$
$$ \alpha = 1.4 $$
$$ \beta = 2

```python
def eqn(x, alpha = 1.4, beta = 2):
    numerator = alpha
    denominator = 1 + np.e ** (-1 * beta * x)
    y = numerator / denominator
    return y
```

```python
x = np.linspace(-4,4,42)
y = eqn(x, alpha = 1.4, beta = 2)
plt.plot(x,y)
```

# Print Formatting and F-Strings

```python
# f strings let you have more control over the output of a print function. For example,
var1 = 10.54
print(f"The variable 'var1' has a value of {var1}.")
```

```python
# f strings work with other data types too
lst = ['Mike', 7]
print(f'{lst[0]} eats {lst[1] * 100} grams of chocolate per day.') # math and indexing work within f-strings
```

```python
# you can also control how many decimal places
pi = 22/7
print(f'The full result is {pi}')
print(f'But rounded to three decimals it looks like {pi:.3f}')
```

# Control Flow

**Comparators**

```python
# special characters that allow you to compare different values and return a Boolean (True or False)
# examples
print(4>5)
print(4<5)
print(4==5)
print(5<=5)
print(5==5)
print(4 != 5)
```

**If Statements**

```python
var = 4
if var == 4: # condition or conditions to evaluate
    print(f'{var} equals 4!') # if the conditional is true, this code will run. Otherwise, nothing happens.
```

```python
var = 5
if var == 4: # condition or conditions to evaluate
    print(f'{var} equals 4!')
```

```python
# can also include alternate conditions using elif and else. each condition is checked in the order it appears
# if one condition is true, the rest of the code block doesn't run
var = 4
if var == 5:
    print('var is 5')
elif var > 5:
    print('var > 5')
else:
    print('var < 5')
print('outside the block')
```

```python
# you can also evaluate multiple conditions at once using and and or operators
if 4==4 and 4 < 10:
    print('example 1')
if 4 == 5 and 4 < 10:
    print('example 2')
if 4 == 5 or 4 < 10:
    print('example 3')
```

**For Loops**

```python
for i in range(0,10): # for item in iterable
    print(i + 1) # each item in the range(0+10) will have one added to it and then will be printed
```

```python
# another example
lst = [2, 'hello', np.linspace(0,1,14)]
for item in lst:
    print(item)
```

```python
# you can combine for loops and if statements
powers = [0] * 10
print(powers)
for i in range(len(powers)):
    if i % 2 == 0 and i > 0:
        print(f'{i} is an even number.')
    if i > 4:
        powers[i] = i ** 2
print(powers)
```

# Measuring Computation Time #

```python
import time
clockStart = time.time()
powers = [0] * 10
print(powers)
for i in range(len(powers)):
    if i % 2 == 0 and i > 0:
        print(f'{i} is an even number.')
    if i > 4:
        powers[i] = i ** 2
print(powers)
comptime = time.time() - clockStart
print(comptime)
```
